```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the Data from the csv file

```python
df = pd.read_csv("Cleaned_Dataset.csv")
```

# Univariate Analysis
1. How Many Unique Job titles are present
2. How many Unique Positions are there

5. What are the Industry types present in the market
6. What is The max, Min and Average salary in Canada

```python
# How Many Unique Job titles are present
print(' How Many Unique Job titles are present')
count_per_job_titles = df['Job Title'].value_counts()
count_per_job_titles = count_per_job_titles.to_dict()
print(count_per_job_titles)
# A plot for the same KPI
labels = list(count_per_job_titles.keys())
counts = list(count_per_job_titles.values())

# Create a pie chart
plt.figure(figsize=(7,7))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Unique Job Titles in Canada')
#plt.axis('equal')
```

```python
# How many unique Positions are there
print(' How Many Unique position are present')
count_per_position = df['Position'].value_counts().reset_index()
# A plot for the same KPI

plt.figure(figsize=(8,8))
plt.title('Count of Positions in the market')
sns.barplot(data=count_per_position,x='Position',y='count')
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_df = df.groupby('Position')['Avg_Salary'].mean()
grouped_df = grouped_df.reset_index()
```

```python
grouped_df.sort_values(by=['Avg_Salary'],ascending=False)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8, 6))  

sns.barplot(grouped_df.head(15),x='Position',y='Avg_Salary')

plt.title('Bar Plot with Custom X-Tick Labels')
plt.xticks(rotation=90)  
plt.show()
```

```python
grouped_df = df.groupby(['City','Position'])['Avg_Salary'].mean().reset_index()
c_p_df_sorted = grouped_df.sort_values(by = 'Avg_Salary')
```

```python
sns.set(style='whitegrid')
fig,axes = plt.subplots(1,2,figsize=(15,15))  

sns.barplot(c_p_df_sorted.head(20),x='City',y='Avg_Salary',hue='Position',ax=axes[0])

axes[0].set_title('Bar Plot with Custom X-Tick Labels')
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=90) 
 

sns.barplot(c_p_df_sorted.tail(20),x='City',y='Avg_Salary',hue='Position',ax=axes[1])

axes[1].set_title('Bar Plot with Custom X-Tick Labels')
axes[1].set_xticklabels(axes[1].get_xticklabels(),rotation=90)  
plt.tight_layout()
plt.show()
```
